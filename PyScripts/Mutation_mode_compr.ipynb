```python
import sys, os
import numpy as np
import matplotlib.pyplot as plt
from importlib import reload
import pandas as pd
import seaborn as sns
```

```python
dt = np.dtype([('patho_mut', np.float), ('Mut_mode', str), ('patho_spp', np.int),
               ('MRCA_time', np.float), ('maxMutNumb', np.float), ('numOfGenes', np.float)])
```

```python
mutDat = pd.read_csv('Mutation_test.csv', delimiter=',', dtype=dt)
```

```python
ax_label = 24
T_label = 28
TicksFS = 22
```

```python
patho_mut = 0.00001
df = mutDat[mutDat['patho_mut'] == patho_mut]
plt.figure(figsize=(12,8))
sns.boxplot(x="patho_spp", y="maxMutNumb", hue="Mut_mode", data=df, palette="Set1")
plt.xticks(fontsize=TicksFS)
plt.yticks(fontsize=TicksFS)
plt.legend(fontsize=TicksFS-2, loc=1)
plt.xlabel("number of pathogen species",fontsize=ax_label)
plt.ylabel("maximal mutation number",fontsize=ax_label)
plt.ylim(top=150)
plt.title("patho.mut. = " + str(patho_mut), fontsize=ax_label)
plt.tight_layout()
plt.savefig(str(patho_mut) + ".maxMutNumb.png")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="patho_spp", y="MRCA_time", hue="Mut_mode", data=df, palette="Set1")
plt.xticks(fontsize=TicksFS)
plt.yticks(fontsize=TicksFS)
plt.legend(fontsize=TicksFS-2, loc=1)
plt.xlabel("number of pathogen species",fontsize=ax_label)
plt.ylabel("generations from MRCA",fontsize=ax_label)
plt.ylim(top=3000)
plt.title("patho.mut. = " + str(patho_mut), fontsize=ax_label)
plt.tight_layout()
plt.savefig(str(patho_mut) + ".MRCA_time.png")
plt.show()
```
